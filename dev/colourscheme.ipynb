# Demo colours

```python
import oscovida
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
N = 20
x = np.linspace(0, 1, N)
for i in range(10):
    color = "C" + str(i)
    y = np.sin(x*2*np.pi)*i/N + 1.2 * i/N
    ax.plot(x, y, 'o-', color=color, label=f'{color}', linewidth=3)
    ax.legend()
    
    
    
fig.savefig('colourscheme.pdf')

    
```
